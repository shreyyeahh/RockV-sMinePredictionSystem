Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and pre-processing


```python
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe() # it describes statistical measures
```

```python
sonar_data[60].value_counts() # count of rock class and mine class
```

```python
#if we didn't have equal no of samples from both classes , then our data would have been imbalanced ---> for that we must use undersampling technique, so our model generalises well
```

```python
sonar_data.groupby(60).mean()
```

```python
#separating the features and the target (data and labels) , its supervised learning problems so we use labels to train the model
```

```python
X = sonar_data.drop(columns=60 , axis = 1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Splitting the data into training and testing data


```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.1 , stratify = Y , random_state = 1) # 0.1 --. we need 10% of data as test data , In the "train_test_split" function, the "stratify" argument is used upon the test data. The sole purpose of the argument is to ensure that the same proportion of examples is preserved across each class of training and testing data , so the equal no of rock and mine sample is there in both training and testing data
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Training the ml model with the help of X_train and Y_train

```python
model = LogisticRegression()
```

```python
model.fit(X_train , Y_train) # our model is now trained on the training data
```

Model Evaluation


accuracy on training data and then on test data


```python
X_train_prediction = model.predict(X_train) # model predicts the value of y based on x_train and X_train_prediction contains the prediction of this model
X_train_prediction
```

```python
training_data_accuracy = accuracy_score(X_train_prediction , Y_train) # calculating the accuracy score of X_train_precition with the actual real corresponding Y_train
training_data_accuracy
```

```python
X_train_prediction = model.predict(X_test) # predicting the value of Y based on test data X_test
testing_data_accuracy = accuracy_score(X_train_prediction , Y_test) # calculating the accuracy score of X_train_precition with the actual real corresponding Y_test
testing_data_accuracy
```

If you have a 2D array [[1, 2], [3, 4], [5, 6]], applying reshape(1, -1) would also result in a 2D array: [[1, 2, 3, 4, 5, 6]]. The elements are flattened and then arranged into a single row.
